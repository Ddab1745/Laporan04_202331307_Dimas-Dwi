```python
#202331307_Dimas Dwi Al Bukhori
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
#202331307_Dimas Dwi Al Bukhori
image = cv.imread("kijang.jpeg")
```

```python
#202331307_Dimas Dwi Al Bukhori
cv.imshow("Gambar kijang", image)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#202331307_Dimas Dwi Al Bukhori
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
edges = cv.Canny(image,75,150) 

cv.imshow("Gambar kijang", image)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#202331307_Dimas Dwi Al Bukhori
fig, axs = plt.subplots(1,2,figsize = (10,10))
axs = axs.ravel()

axs[0].imshow(gray, cmap="gray")
axs[0].set_title("Gambar Asli")

axs[1].imshow(edges, cmap="gray")
axs[1].set_title("Gambar Edges")
```

```python
#202331307_Dimas Dwi Al Bukhori
lines = cv.HoughLinesP(edges,1,np.pi/180,30,maxLineGap=250)
image_line = image.copy()
```

```python
#202331307_Dimas Dwi Al Bukhori
for line in lines:
    x1,y1,x2,y2 = line[0]
    cv.line(image_line, (x1,y1),(x2,y2),(100,8,255),1)
```

```python
#202331307_Dimas Dwi Al Bukhori
fig, axs = plt.subplots(1,2,figsize = (10,10))
axs = axs.ravel()

axs[0].imshow(edges, cmap="gray")
axs[0].set_title("Gambar Edges")

axs[1].imshow(image_line, cmap="gray")
axs[1].set_title("Gambar Line")
```

```python
#202331307_Dimas Dwi Al Bukhori
import numpy as np
import matplotlib.pyplot as plt
import cv2 
img_jalan = cv2.imread("jalan.jpg",0)
rows, cols = img_jalan.shape
print("IMG SHAPE: ", img_jalan.shape)
```

```python
#202331307_Dimas Dwi Al Bukhori
M = cv2.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0), 45,1)
img_putar = cv2.warpAffine(img_jalan, M, (cols,rows))

fig, axs = plt.subplots(1,2,figsize = (10,5))
axs = axs.ravel()

axs[0].imshow(img_jalan, cmap="gray")
axs[0].set_title("Gambar Asli")

axs[1].imshow(img_putar, cmap="gray")
axs[1].set_title("Gambar Putar")

for a in axs:
    a.axis('off')

plt.tight_layout()
plt.show()
```

```python
#202331307_Dimas Dwi Al Bukhori
from skimage import io,transform

img_jalan2 = io.imread("jalan.jpg")

rotated = transform.rotate(img_jalan2, -15, resize=False)
rotated2 = transform.rotate(img_jalan2, -15, resize=True)

fig, axs = plt.subplots(1,3,figsize = (10,5))
axs = axs.ravel()

axs[0].imshow(img_jalan2, cmap="gray")
axs[0].set_title("Gambar Asli")

axs[1].imshow(rotated, cmap="gray")
axs[1].set_title("Gambar Putar 1")

axs[2].imshow(rotated2, cmap="gray")
axs[2].set_title("Gambar Putar 2")

for a in axs:
    a.axis('off')

plt.tight_layout()
plt.show()
```
